# brightness and contrast

```java
%mavenRepo clojars https://clojars.org/repo/
%maven origami:origami:4.1.1-8

import org.opencv.imgcodecs.Imgcodecs;
import origami.Origami;
import org.opencv.core.Mat;
import org.opencv.core.Core;
Origami.init();
```

```java
Mat marcel = Imgcodecs.imread("../data/marcel2019.jpg", Imgcodecs.IMREAD_REDUCED_COLOR_8);
Origami.matToBufferedImage(marcel)
```

## Change contrast

```java
Mat contrast1 = new Mat();
marcel.convertTo(contrast1,-1,2,0);
Origami.matToBufferedImage(contrast1);
```

```java
Mat contrast2 = new Mat();
marcel.convertTo(contrast2,-1,0.5,0);
Origami.matToBufferedImage(contrast2);
```

## Change brightness

```java
Mat contrast3 = new Mat();

marcel.convertTo(contrast3, -1, 1, 100);
Origami.matToBufferedImage(contrast3);
```

```java
Mat contrast4 = new Mat();
marcel.convertTo(contrast4, -1, 1, 50);
Origami.matToBufferedImage(contrast4);
```

```java
Mat contrast4 = new Mat();
marcel.convertTo(contrast4, -1, 1, -50);
Origami.matToBufferedImage(contrast4);
```

# Concatenation

```java
Mat target = new Mat();
List<Mat> src = new ArrayList();
int[] brightnessValues = new int[]{-100,-50,10,50,100};
for(int b : brightnessValues) {
    Mat h = new Mat();
    marcel.convertTo(h, -1, 1, b);
    src.add(h);
}
Core.hconcat(src, target);
Origami.matToBufferedImage(target);
```

```java
Mat target = new Mat();
List<Mat> src = new ArrayList();
double[] contrastValues = new double[]{0.2,0.5,1,1.5,2};
for(double b : contrastValues) {
    Mat h = new Mat();
    marcel.convertTo(h, -1, b, 0);
    src.add(h);
}
Core.hconcat(src, target);
Origami.matToBufferedImage(target);
```
