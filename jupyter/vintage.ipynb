```python
# from
# https://medium.com/ai-simplified/create-your-own-instagram-filters-using-python-6d8bd3a86d4
import cv2
import numpy as np
from matplotlib import pyplot as plt

im = cv2.imread('../data/house.jpeg')
rows, cols = im.shape[:2]
```

```python
# Create a Gaussian filter
kernel_x = cv2.getGaussianKernel(cols,200)
kernel_y = cv2.getGaussianKernel(rows,200)

print(kernel_x[0])
print(kernel_y[0])

t = kernel_x.T
print(t[0])
```

```python
kernel = kernel_y.dot(kernel_x.T)
```

```python
print(kernel[4])
```

```python


print(kernel_y.shape)
print(kernel_x.shape)
print(kernel_x.T.shape)
print(kernel.shape)
filter = 255 * kernel / np.linalg.norm(kernel)
# print(filter)
vintage_im = np.copy(im)
```

```python
# for each channel in the input image, we will apply the above filter
for i in range(3):
    vintage_im[:,:,i] = vintage_im[:,:,i] * filter
```

```python
plt.imshow(im)
plt.show()
plt.imshow(vintage_im)
plt.show()
```
